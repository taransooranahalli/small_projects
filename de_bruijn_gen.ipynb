```python
import networkx as nx
import matplotlib.pyplot as plt

def de_bruijn_graph(k):
    nodes = [bin(i)[2:].zfill(k) for i in range(2**k)]
    graph = nx.DiGraph()
    graph.add_nodes_from(nodes)

    for node in nodes:
        graph.add_edge(node, node[1:] + '0')
        graph.add_edge(node, node[1:] + '1')
    return graph

k = 4 
graph_16 = de_bruijn_graph(k)


pos = nx.circular_layout(graph_16)
nx.draw(graph_16, pos, with_labels=True, node_size=1000, node_color="purple", font_size=10, arrowsize=20)
plt.title(f"De Bruijn Graph with {len(graph_16.nodes)} Vertices (k={k})")
plt.show()


# print("Edges of the De Bruijn graph:")
# for edge in graph_16.edges():
#     print(edge)
```
